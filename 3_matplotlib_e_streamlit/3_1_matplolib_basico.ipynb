# 3.1 Exploração inicial do Matplotlib

O objetivo é fazer um gráfico de barras simples com Matplotlib

```python
# Importar a biblioteca matplotlib

import matplotlib.pyplot as plt
```

```python
# Criação de uma figura (fig) e um conjunto de eixos (ax)

fig, ax = plt.subplots()
```

```python
# Geramos dados de exemplo para alimentar o gráfico

contagem_medicamentos = {
    'metformina': 100,
    'DDP4': 30,
    'i-SGLT2': 55,
    'analogos GLP1': 25
}

medicamentos = list(contagem_medicamentos.keys())
# for each in medicamentos:
#     medicamentos[medicamentos.index(each)] = each.capitalize()
contagens = list(contagem_medicamentos.values())
```

```python
# Definir as labels e cores

bar_labels = ['Metformina', 'iDDP4', 'i-SGLT2', 'analogos GLP1']
bar_colors = ['tab:green', 'tab:blue', 'tab:red', 'tab:orange']
```

```python
# Defininmos um grafico de barras (ax.bar) e passamos os dados, labels e cores
ax.bar(
    medicamentos, # x=medicamentos,
    contagens, # height=contagens,
    label=bar_labels,
    color=bar_colors
)
fig
```

```python
# definimos o nome dos eixos, título e legenda

ax.set_ylabel('Contagem de medicamentos prescritos')
ax.set_xlabel('Tipo de medicamento')

ax.set_title('Fornecimento de medicamentos por tipo e cor')

ax.legend(title='Cor do medicamento')
```

```python
# mostrar o gráfico (se estivessemos num script de Python .py seria plt.show() em vez de apenas fig ou ax.figure)

fig
```

```python
# adicionar os valores no topo de cada barra
for i in range(len(contagens)):
    ax.text(i, contagens[i] + 2, str(contagens[i]), ha='center', fontdict={'size': 12})
fig
```

```python
# para acomodar melhor os elementos do gráfico, podemos definir os limites dos eixos
ax.set_ylim(0, max(contagens) * 1.1)
fig
```

```python
# Vamos usar os mesmos dados para fazer um pie chart
# fazemos uma nova figura e novos eixos
fig_2, ax_2 = plt.subplots()
```

```python
# mudar as cores para corresponder ao gráfico de barras anterior
ax_2.pie(
    contagens,
    labels=medicamentos,
    colors=bar_colors,
    autopct='%1.1f%%',
    startangle=90,
    #textprops={'fontsize': 16}
)
ax_2.set_title('Distribuição das frutas')
fig_2
```

```python
# adicionar o total n =  no canto do gráfico
total = sum(contagens)
ax_2.text(-1.5, 1, f'n = {total}', fontdict={'size': 14})
fig_2
```
